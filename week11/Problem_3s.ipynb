## Question 1:  Parsing HTML file using Beautiful Soup

The above cell contains html text written to `countries.html` that contains a table that with all the countries in the world and their respective capital cities. Using Beautiful Soup, complete the function `countries_map()` and return a dictionary that maps each capital to its respective country. For example, by observing the html above, it is evident that the return dictionary should be formatted as such: `{'Kabul':'Afghanistan', 'Tirana': 'Albania', 'Algiers': 'Algeria', ...}`.

```python
def countries_map(html_file):
    """
    Construct a dictionary mapping the country to capital 
    in the above countries.html file
    
    Parameters
    ----------
    html_file: a string that represents the filepath to the html file to be parsed
    
    Returns
    -------
    A dictionary
    """
    
    # YOUR CODE HERE

    html = open(html_file, 'r')
    soup = BeautifulSoup(html, 'lxml')
    tables = soup.find_all('table')
    
    country = []
    captail = []
    
    countries = {}
    
    i = 0
    for table in tables:

        for tdx in table.find_all('td'):
            if i % 2 == 0:
                country.append(tdx.contents[0])
            else:
                captail.append(tdx.contents[0])
            i += 1

    countries = dict(zip(captail, country))
    html.close()
    return countries
```

```python
countries = countries_map('countries.html')
assert_equal(countries['Kabul'], 'Afghanistan')
assert_equal(countries['Washington, D.C.'], 'United States of America')
assert_equal(countries['Buenos Aires'], 'Argentina')
assert_equal(len(countries), 197)
```

In this question, we will be implementing a simple parsing function, that will return to us the most powerful pokemon defined **greater than or equal to above a certain threshold**. We will quantify the power of a pokemon, in terms of the `total` attribute for each pokemon (i.e. the 4th column value). Finish the function below, `get_pokemon_above_threshold` that returns a list of pokemon names (list of strings) and takes in 2 function parametesr: `threshold`, an integer and `html_file`, a string that represents the file path of the html file to be parsed.

- For example, given the first 5 pokemon and if the function call was `get_pokemon_above_threshold(350)`, then `[Ivysaur, Venusaur, Charmeleon]` would be returned since these are the 3 pokemon with totals **greater than or equal to** 350.
- Note that `Venusaur` is only added once, even though there is a second version of Venusaur with a higher total. Therefore, do not double count the same pokemon with a `Mega` prefix.

```python
def get_pokemon_above_threshold(threshold, html_file):
    '''    
    Return a list of pokemon names above the threshold
    
    Parameters
    ----------
    threshold: an int
    html_file: a string that represents the filepath to the html file to be parsed
    
    Returns
    -------
    A list of strings
    '''
    
    # YOUR CODE HERE
    html = open(html_file)
    soup = BeautifulSoup(html, 'lxml')
    tb = soup.find('table').tbody

    total = []
    pokemon = []
    
    for row in tb.find_all('tr'):   
        tdx = [val for val in row.find_all('td')]
        if (tdx[1].a.contents[0] not in pokemon) & \
        (int(tdx[3].contents[0].strip()) >= threshold):
            pokemon.append(tdx[1].a.contents[0].strip())
            
#             total.append(tdx[3].contents[0].strip()
            
    return pokemon

    
```
